## A) Recommendation system based on Average Weight


The Simple Recommender system provides broad movie suggestions to all users by considering overall movie popularity and occasionally genre preferences. The underlying principle of this recommender lies in the notion that movies with higher popularity and critical acclaim are more likely to be appreciated by the average audience. Notably, this model does not deliver personalized recommendations tailored to individual users.

The implementation of this recommendation model is exceptionally straightforward. It involves sorting movies based on their ratings and popularity, then presenting the top movies from this sorted list. Additionally, the system allows users to specify a particular genre, enabling them to obtain the top 20 movies in that specific genre. By utilizing ratings, the recommendation system identifies and showcases the most popular movies within the chosen genre, enhancing user experience and ensuring that movie suggestions align with their genre preferences.

I utilize TMDB Ratings as the basis for our Top Movies Chart, employing IMDB's weighted rating formula for construction. Mathematically, this formula is represented as:

Weighted Rating $(W) = ((v/v+m) * R) + ((m/v+m) * C)$

where,

- v is the number of votes for the movie
- m is the minimum votes required to be listed in the chart
- R is the average rating of the movie
- C is the mean vote across the entire dataset.

The subsequent step involves determining a suitable value for 'm,' the minimum votes needed for a movie to be featured in the chart. We opt for the 85th percentile as our threshold. In simpler terms, a movie must have more votes than at least 85% of the movies in the list to be included in our charts.

Here we will construct our comprehensive Top 250 movies in the chart and develop a function to create charts tailored to specific genres.




```python
# Now we will calculate all the components of the weighted rating

v = movies_clean_df['vote_count']
R = movies_clean_df['vote_average']
C = movies_clean_df['vote_average'].mean()
m = movies_clean_df['vote_count'].quantile(0.85)
```

```python
movies_clean_df['weighted_average'] = ((R*v)+(C*m))/(v+m)
movies_clean_df.head()
```

For the recommendation system based on average, the particular columns which we need are: genre, popularity, release_date, vote_average, vote_count and weighted_average.

To resolve this problem we can make a new recommendation system where weighed average and popularity both are given 50-50 priority. But the value of weighted average is in the range of 0 to 10 and the value of popularity is the range of 100 to 1000. 

Therefore, their is huge difference in the range of both the feature. So to obtain 50% priority of both the feature, we need to normalize them.

Now we will construct our function that builds charts for particular genres.

```python
# Here, we are processing the 'genres' column of the DataFrame `movie_scored_df` to extract individual genres for each movie.

s = movie_scored_df.apply(lambda x: pd.Series(x['genres']),axis=1).stack().reset_index(level=1, drop=True)
s.name = 'genre'
# Dropping genre colunm from original movie scored dataframe and modified DataFrame is joined with the Series s (which contains the individual genres) on the DataFrame's index. 
gen_movie_scored_df = movie_scored_df.drop('genres', axis=1).join(s)
```

```python
def build_chart(genre, percentile=0.85):
    # Filter movies based on the specified genre
    df = gen_movie_scored_df[gen_movie_scored_df['genre'] == genre]
    
    # Select specific columns from the DataFrame
    selected_columns = ['title', 'release_date', 'score', 'weighted_average', 'popularity']
    df = df[selected_columns]
    
    # Sort movies based on the 'score' column in descending order
    df = df.sort_values('score', ascending=False).head(250)
    
    return df  # Return the modified DataFrame containing recommended movies
```

```python
recommended_movies = build_chart('Action')  # Replace 'Action' with the desired genre
recommended_movies.head(25)
```

```python
recommended_movies = build_chart('Romance')  
recommended_movies.head(25)
```

The recommender system we constructed in the previous section comes with notable limitations. One major drawback is that it offers the same set of recommendations to all users, regardless of their unique preferences. For example, if a person is a fan of romantic movies but dislikes action films, the movies suggested in our Top 15 Chart might not align with their tastes. Even if this individual explores our genre-specific charts, they still won't receive personalized recommendations tailored to their preferences. So to resolve this issue, we will construct a content-based recommendation system.

## B) Content Based Recommendation System

To enhance the personalization of our recommendations, we will develop an engine that calculates similarities between movies using specific criteria. This engine will suggest movies that closely resemble a particular film that a user enjoyed. This approach, utilizing movie metadata or content, is commonly referred to as Content-Based Filtering.

We will create Content-Based Recommenders focusing on:

- Movie Overviews 
- Movie Cast
- Director
- Keywords
- Genre

```python
movie_cb = movies_clean_df[['id','title','genres','keywords','overview','cast','director']]
movie_cb.head(10)
```

```python
# In the above dataframe we need to remove space between a specfic tag so that it appears to be a single word.

def collaspe(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# Assuming collaspe() and collapse() are our functions to process the columns

# Use .loc[] to modify the DataFrame directly
movie_cb.loc[:,'cast'] = movie_cb['cast'].apply(collaspe)
movie_cb.loc[:,'director'] = movie_cb['director'].apply(collaspe)
movie_cb.loc[:,'genres'] = movie_cb['genres'].apply(collaspe)
movie_cb.loc[:,'keywords'] = movie_cb['keywords'].apply(collaspe)

movie_cb.head(10)
```

```python
# Spliting string of words in overview into individual tags.
movie_cb['overview'] = movie_cb['overview'].apply(lambda x:x.split()) 
```

```python
# Now we will create a collection of tags from the columns of genres, overview, keywords, cast and director.

movie_cb['tags'] = movie_cb['overview']+movie_cb['genres']+movie_cb['keywords']+movie_cb['cast']+movie_cb['director']
tagged_movie = movie_cb.drop(columns=['overview','genres','keywords','cast','director'])
tagged_movie['tags'] = tagged_movie['tags'].apply(lambda x:" ".join(x))
tagged_movie.head(10)
```

```python
# Changing all the tags into lowercase.
tagged_movie['tags']=tagged_movie['tags'].apply(lambda x:x.lower())
tagged_movie.head()
```

```python
tagged_movie['title'].values
```

```python
# The tags we made might have similar words like dance, dancing etc. To remove the problem of stemming we will apply natural language tool kit (nltk).

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
tagged_movie['tags']=tagged_movie['tags'].apply(stem)
```

```python
tagged_movie['tags'][0]
```

```python
# Model Building

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') # Taking hyperparameters for maximum word limit 5000 and remove stopping words of english.
```

```python
# Converting each tags of particular movie into set of array and forming a matrix.

vectors = cv.fit_transform(tagged_movie['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(vectors)
tagged_movie['similarity'] = similarity_matrix.tolist()
```

```python
similarity_matrix.shape
```

```python
# Building recommendation system on the basis of similarity of the arrays by calculating he minimum distance between two pairs of array.

def recommend(movie):
    if movie not in tagged_movie['title'].values:
        print(f"Movie '{movie}' not found in the database.")
        return
    index = tagged_movie[tagged_movie['title'] == movie].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1]) # Enumerate function hold the index position for each movie id
    for i in distance[1:6]:
        print(tagged_movie.iloc[i[0]].title)
```

```python
recommend('Avatar') # Replace 'Avatar' with desired movie title.
```

```python
recommend('Batman Begins')
```

```python
recommend('Harry Potter')
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
